### **Задача 1:**  
На чемодане есть кодовый замок из 4х цифр. Цифры в коде не должны повторяться. Порядок цифр важен: то есть комбинации {1234}, {2341}
и {4321} — это разные коды.  
Сколько комбинаций кода можно придумать?  

**Решение для задачи 1: (вариант 1)** 
Чтобы определить количество комбинаций кода, можно использовать принципы комбинаторики.  
Для первой цифры в коде у нас есть 10 возможностей (от 0 до 9), так как мы можем использовать любую цифру.  
После выбора первой цифры, у нас остаются 9 возможных цифр для выбора второй цифры. Поскольку цифры не должны повторяться, это означает, что для второй цифры у нас остается только 9 вариантов выбора.  
Аналогично, для третьей цифры остается 8 возможностей, и для четвертой цифры — 7 возможностей.  
Таким образом, общее количество комбинаций кода можно определить умножив количество возможностей для каждой цифры:

10 * 9 * 8 * 7 = 5040 комбинаций кода.

Таким образом, можно придумать 5040 различных комбинаций кода для данного чемодана с 4-значным кодовым замком.  

**Решение для задачи 1: (вариант 2)** 
Количество комбинаций кода можно также рассчитать с использованием понятия перестановки. Поскольку порядок цифр важен, мы будем использовать формулу для перестановок.  
Формула для перестановок из n элементов (в данном случае цифр) по k элементов (длина кода) выглядит следующим образом:  
P(n, k) = n! / (n - k)!  
Где n! обозначает факториал числа n, что означает произведение всех натуральных чисел от 1 до n.  
В нашем случае у нас 10 возможных цифр (от 0 до 9) и мы выбираем 4 цифры для кода:  

P(10, 4) = 10! / (10 - 4)! = 10! / 6! = (10 * 9 * 8 * 7 * 6!) / 6! = 10 * 9 * 8 * 7 = 5040

Таким образом, мы получаем ту же самую сумму 5040 различных комбинаций кода, что и в предыдущем решении.  

**Ответ для задачи 1: 5040**   

```python
from math import factorial

factorial(10)/factorial(10-4)
```

### **Задача 2:**  
Сколько мелодий из 3 нот можно составить из 7 нот одной октавы? Порядок учитывается: «до, ре, ми» и «ми, ре, до» — это разные мелодии.  

**Решение для задачи 2:** 
Чтобы определить количество мелодий из 3 нот, которые можно составить из 7 нот одной октавы с учетом порядка, мы можем использовать принцип размещений.  
Формула для размещений из n элементов по k элементов выглядит следующим образом:  
A(n, k) = n! / (n - k)!  
Где n! обозначает факториал числа n, что означает произведение всех натуральных чисел от 1 до n.  
В данном случае у нас 7 возможных нот, и мы выбираем 3 ноты для мелодии:  
A(7, 3) = 7! / (7 - 3)! = 7! / 4! = (7 * 6 * 5 * 4!) / 4! = 7 * 6 * 5 = 210

Таким образом, мы можем составить 210 различных мелодий из 3 нот, используя 7 нот одной октавы с учетом порядка.  

**Ответ для задачи 2: 210**   

```python
from math import factorial

factorial(7)/factorial(7-3)
```

### **Задача 3:**  
Сколько музыкальных плейлистов длиной в 5 песен можно составить из 10 треков? Плейлисты с одинаковыми композициями и разным порядком песен считаются разными.  

**Решение для задачи 3:** 
Чтобы определить количество музыкальных плейлистов длиной в 5 песен, которые можно составить из 10 треков с учетом разного порядка песен, мы можем использовать принцип размещений.  
Формула для размещений из n элементов по k элементов выглядит следующим образом:  
A(n, k) = n! / (n - k)!  
Где n! обозначает факториал числа n, что означает произведение всех натуральных чисел от 1 до n.  
В данном случае у нас 10 возможных треков, и мы выбираем 5 треков для плейлиста:  
A(10, 5) = 10! / (10 - 5)! = 10! / 5! = (10 * 9 * 8 * 7 * 6 * 5!) / 5! = 10 * 9 * 8 * 7 * 6 = 30240

Таким образом, мы можем составить 30240 различных музыкальных плейлистов длиной в 5 песен, используя 10 треков с учетом разного порядка песен.  

**Ответ для задачи 3: 30240**   

```python
from math import factorial

factorial(10)/factorial(10-5)
```
