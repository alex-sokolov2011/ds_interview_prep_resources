Чтобы максимизировать ожидаемую прибыль, если на первом броске выпадает сумма меньше 3.5 доллара, мы должны выбрать второй бросок, так как ожидаемый выигрыш от второго броска будет выше. $E(при\,тактике:\,первый < 3.5) = ∑_{i=1}^{6} P(X=i) \cdot i = \frac{E(при\,двух\,бросках)}{2} + \frac{E(при\,одном\,броске)}{21} \cdot 15 +  = \frac{3.5}{2} +  \frac{3.5}{21}\cdot 15 = 1.75 + 2.5 = 4.25$

### Задача 2. Как можно проверить решение задачи 1 с помощью метода Монте-Карло

```python
import random
import matplotlib.pyplot as plt

# Количество граней кубика
num_sides = 6

# Граница, при которой решается перебросить кубик
threshold = 3.5

# Функция для броска кубика
def throw_dice():
    """ Симуляция броска кубика """
    return random.randint(1, num_sides)

# Функция для расчета выигрыша с учетом порога
def calculate_gain(first_throw, threshold):
    """ Расчет выигрыша в зависимости от порога """
    if first_throw < threshold:
        # Переброс, если выпало меньше порога
        return throw_dice()
    else:
        # Иначе сохраняем результат первого броска
        return first_throw

# Симуляция игры с возможностью переброса кубика
def play_game(threshold):
    """ Симуляция игры с возможностью переброса кубика в зависимости от порога """
    first_throw = throw_dice()
    return calculate_gain(first_throw, threshold)

# Функция для проведения множества игр и расчета среднего выигрыша
def simulate_games(num_games, threshold):
    results = [play_game(threshold) for _ in range(num_games)]
    average_result = sum(results) / len(results)
    return average_result

# Проведение симуляции и расчет среднего значения для разных порогов
thresholds = [i + 0.5 for i in range(num_sides + 1)]
average_results = [simulate_games(100000, threshold) for threshold in thresholds]
```

```python
for t, a in zip(thresholds, average_results):
    print(f'При пороге: {t} матожидание == {a}')
```

```python
# Построение графика зависимости среднего выигрыша от порога
plt.plot(thresholds, average_results, marker='o')
plt.title('Зависимость математического ожидания выигрыша от порога')
plt.xlabel('Порог для переброса')
plt.ylabel('Средний выигрыш')
plt.grid(True)
plt.show()
```

**Решение задачи 3: вариант 2**

Вспомним чему равно было матожидание при выборе тактики перебрасываем если выпало меньше 3.5  

$E(при\,тактике:\,первый < 3.5) = \frac{E(при\,двух\,бросках)}{2} + \frac{E(при\,одном\,броске)}{21} \cdot 15 +  = \frac{3.5}{2} +  \frac{3.5}{21}\cdot 15 = 1.75 + 2.5 = 4.25$  

Попробуем обобщить формулу в зависимости от порога $x$ при котором мы перебрасываем:  
$E(при\,тактике:\,первый < 3.5) = \frac{3.5}{6} \cdot x + \frac{(21-\frac{x*(x+1)}{2}}{6}$

Проверим что при x=1 и 6 получаем 3.5 , а при x=3 получаем 4.25 (выведим график функции)

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию
def f(x):
    return 3.5*x/6 + (21 - x*(x+1)/2)/6

# Создаем массив значений x от 0 до 6
x = np.linspace(0, 6, 300)

# Вычисляем значения функции в точках x
y = f(x)

# Строим график
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='y = 3.5*x/6 + (21 - x*(x+1)/2)/6')
plt.title('График функции')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

Если умножить на 6, то сможем упростить до читаемого в этом редакторе вида  
$3.5\cdot x + 21-\frac{x*(x+1)}{2}$  

Это квадратное уравнение и нам надо найти его максимум, поэтому упрощаем дальше и берем производную:  
$-0.5x^2 + 3x + 21.0$ - это парабола ветвями вниз

Проиводная: $-x + 3 = 0$  
$x == 3$  
При этом $x$ матожидание максимальное и равно 4.25

**Решение задачи 3: вариант 3**

Пусть $A$ – первый бросок, $B$ – повторный бросок. $A, B \in \{1, 2, 3, 4, 5, 6\}$ 

Выберем стратегию:

Рассмотрим $A = 3$. Вероятность выбросить вторым броском меньше ($P(B<A | A=3)$) = $\frac{2}{6}$.  
А больше ($P(B>A | A=3)$) = $\frac{3}{6}$. Т. е., вероятнее всего, результат будет улучшен при повторном броске.

Аналогично для $A = 4$: $P(B<A | A=4) = \frac{3}{6}$, $P(B>A | A=4) = \frac{2}{6}$. Т. е., вероятнее всего, результат будет ухудшен при повторном броске.

На основе этого получаем следующую стратегию: если выбросили 4 или больше, то НЕ перебрасываем. Иначе – перебрасываем.

**Решение задачи 3: вариант 4**  
Пороговое значение можно точнее определить посредством оценки мат. ожидания первого броска (выше мы его рассчитали и оно равно 3.5).  
Если мы первым броском выбросили значение больше мат. ожидания, то, перебрасывая, в среднем, мы только ухудшим результат.  

Посчитаем мат. ожидание выбранной стратегии:

Если на первом шаге выпала 1, мат. ожидание этой части: $\text{last}(1,1) \times \frac{1}{36} + \text{last}(1,2) \times \frac{1}{36} + ... + \text{last}(1,6) \times \frac{1}{36} = \frac{21}{36}$, где $\text{last}(A, B) = B$ – значение последнего броска.

Аналогично для 2 и 3.

Если на первом шаге выпала 4, мат. ожидание этой части: $4 \times \frac{1}{6} = \frac{4}{6}$, т. к. не перебрасываем кубик.

Аналогично для 5 и 6.

Теперь все складываем и получаем $E = \frac{21}{36} + \frac{21}{36} + \frac{21}{36} + \frac{4}{6} + \frac{5}{6} + \frac{6}{6} = 4.25$

Итого: мат. ожидание оптимальной стратегии равно 4.25.
