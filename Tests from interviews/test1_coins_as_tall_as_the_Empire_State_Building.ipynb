**Решение для задачи 1:**  

Выдержка из [википедии](https://ru.wikipedia.org/wiki/%D0%AD%D0%BC%D0%BF%D0%B0%D0%B9%D1%80-%D1%81%D1%82%D0%B5%D0%B9%D1%82-%D0%B1%D0%B8%D0%BB%D0%B4%D0%B8%D0%BD%D0%B3) о Эмпайр-стейт-билдинг:

В здании 102 этажа, высота его основной части — 381,3 метра. С телевизионной башней, надстроенной в 1950-е годы вместо причальной мачты для дирижаблей, оно достигает 443 метров, а ширина у основания — около 140 м. Коммерческие площади занимают первые 85 этажей здания (257 211 м²). Остальные 16 этажей — это надстройка в стиле ар-деко, на 102 этаже расположена смотровая площадка.  

Встает вопрос - какая высота первых 85 этажей без надстройки?

Из схемы видно, что на первые 85 этажей приходится примерно 300 метров. Следовательно высота одного этажа в среднем ~ 3.529 метра.  
Но надо учесть, что между этажами есть перекрытия. И нужно узнать их примерную толщину.  Из фотографий строительства видно, что здание построено с помощью железобетонных конструкций.  

Поиск дополнительной информации дал:  
В те годы толщина большинства перекрытий составляла 250-300 мм, но достигала и 450-500 мм в верхней части здания (для придания дополнительного веса).  

Мы можем принять, что средняя высота перекрытия равна максимальному значению 500 мм, тем самым средняя высота этажа без перекрытия 3.029 метра. При этом мы будем уверенны, что если при таком условии нам удастся разместить монеты в помещении, то в реальности высота помещения без перекрытия будет выше установленной.  

Теперь посчитаем сколько стопок нам необходимо разместить в помещении:  
443.2/3.029 = 146.3189 ~ 147 стопок.

Теперь надо узнать какую площадь занимает одна монета номиналом в один пенс. По информации из [википедии](https://ru.wikipedia.org/wiki/%D0%91%D1%80%D0%B8%D1%82%D0%B0%D0%BD%D1%81%D0%BA%D0%B8%D0%B9_%D0%BF%D0%B5%D0%BD%D0%BD%D0%B8) диаметр монеты менялся и максимальным он был в размере примерно 30 мм, или с радиусом R=15мм=0.015м  

Квадрат описывающий монету диаметром 30 мм имеет сторону равную 2R и площадь 4R^2 = 0.0009 квадратных метров:  

147 стопок будут занимать площадь = 147*0.0009 = 0.1323 м^2  

**Ответ для задачи 1: Да, конечно, поместились бы так-как площадь этих монет высотой от пола до потолка занимала бы менее 0.14 квадратного метра**

Аналогичным образом можно сделать расчет какой высоты должна быть стопка монет диметром 30 мм чтобы разместить ее на одном квадратном метре в помещении 3 метра высотой?

**Дополнение к ответу для задачи 1:**  
Какой должен быть диаметра монеты, чтобы они занимали один квадратный метр?  

Если на одном квадратном метре должно разметиться 147 стопок монет, то площадь занимаемая монетой будет равна:  
1/147 = 0.0068 м^2  

4R^2 = 0.0068 = $S_m$  
отсюда следует что радиус монеты должен быть равен = √($S_m$/4) = 0.041 метра = 41 мм. А диаметр соответсвенно 82 мм.  

То есть однозначно можно разместить эти монеты даже на одном квадратном метре в помещении с высотой потолков не менее средней в этом здании, так как даже представить себе монету диаметром в 82 мм достаточно сложно.
